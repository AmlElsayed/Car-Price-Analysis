# 1- What are different car brands in the dataset?

```python
unique_brands=df['Brand'].unique()
unique_brands
```

# 2- How many cars are there in the dataset?

```python
total_cars=df['Car_ID'].count()
total_cars
```

# 3- What is average price for all cars?


```python
average_price=df['Price'].mean()
average_price
```

# 4- Which car has the highest mileage?

```python
df[df['Mileage']==df['Mileage'].max()][['Brand','Model']]
```

# 5- What is the average Kilometres driven for each brand ?

```python
avg_kms_per_brand=df.groupby('Brand')['Kilometers_Driven'].mean()
avg_kms_per_brand
```

# 6- Which fuel type is most common among the cars?

```python
common_fuel_type=df['Fuel_Type'].mode()[0] 
common_fuel_type
```

# 7- What is the maximum power output of a car in the dataset?

```python
max_power=df['Power'].max()
max_power
```

# 8- How many cars were manufactured each year?-----

```python
cars_per_year=df.groupby('Year')['Car_ID'].count()
cars_per_year
```

# 9- What is average number of seats for each brand?

```python
avg_seats_per_brand=df.groupby('Brand')['Seats'].mean().astype(int)
avg_seats_per_brand
```

# 10- Which car has the highest engine capacity?

```python
max_engine_car=df[df['Engine']==df['Engine'].max()][['Brand','Model']]
max_engine_car
```

# 11- What is the most common owner type among the car?

```python
common_owner_type=df['Owner_Type'].mode()[0]
common_owner_type
```

# 12- What is the range of prices for cars in the dataset?

```python
price_range=(df['Price'].min(),df['Price'].max())
price_range
```

# 13- How many unique car models are there in the dataset?

```python
unique_models_count=df['Model'].nunique()
unique_models_count
```

# 14- What is the average mileage for each fuel type?

```python
avg_mileage_per_fuel_type=df.groupby('Fuel_Type')['Mileage'].mean()
avg_mileage_per_fuel_type
```
